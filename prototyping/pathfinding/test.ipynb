```python
from pathfinder import Pathfinder
from grid import Grid
```

```python
pathfinder = Pathfinder()
grid = Grid(256, 128)
```

```python
pathfinder.clear_graph()
```

```python
for row_i in range(grid.height):
    for col_i in range(grid.width):
        vertex_index = grid.cell_1d_index(row_i, col_i)
        pathfinder.add_vertex(vertex_index)
```

```python
pathfinder.graph.adjacency_list[0]
```

```python
for row_i in range(grid.height):
    for col_i in range(grid.width):
        current_index = grid.cell_1d_index(row_i, col_i)
        up = grid.cell_1d_index(row_i-1, col_i)
        down = grid.cell_1d_index(row_i+1, col_i)
        left = grid.cell_1d_index(row_i, col_i-1)
        right = grid.cell_1d_index(row_i, col_i+1)
        up_left = grid.cell_1d_index(row_i-1, col_i-1)
        up_right = grid.cell_1d_index(row_i-1, col_i+1)
        down_left = grid.cell_1d_index(row_i+1, col_i-1)
        down_right = grid.cell_1d_index(row_i+1, col_i+1)

        if not grid.out_of_bounds(row_i-1, col_i):
            if not pathfinder.edge_in_graph(up, current_index):
                pathfinder.add_edge(up, current_index, 2)

        if not grid.out_of_bounds(row_i+1, col_i):
            if not pathfinder.edge_in_graph(down, current_index):
                pathfinder.add_edge(down, current_index, 2)

        if not grid.out_of_bounds(row_i, col_i-1):
            if not pathfinder.edge_in_graph(current_index, left):
                pathfinder.add_edge(current_index, left, 2)

        if not grid.out_of_bounds(row_i, col_i+1):
            if not pathfinder.edge_in_graph(current_index, right):
                pathfinder.add_edge(current_index, right, 2)

        if not grid.out_of_bounds(row_i-1, col_i-1):
            if not pathfinder.edge_in_graph(up_left, current_index):
                pathfinder.add_edge(up_left, current_index, 3)

        if not grid.out_of_bounds(row_i-1, col_i+1):
            if not pathfinder.edge_in_graph(up_right, current_index):
                pathfinder.add_edge(up_right, current_index, 3)
        
        if not grid.out_of_bounds(row_i+1, col_i-1):
            if not pathfinder.edge_in_graph(down_left, current_index):
                pathfinder.add_edge(down_left, current_index, 3)

        if not grid.out_of_bounds(row_i+1, col_i+1):
            if not pathfinder.edge_in_graph(down_right, current_index):
                pathfinder.add_edge(down_right, current_index, 3)
```
